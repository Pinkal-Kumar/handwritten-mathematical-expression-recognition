```python
import os
from PIL import Image 
import json

folder_images = "off_image_train"
size_images = dict()
height_list = []
width_list = []
for dirpath, _, filenames in os.walk(folder_images):

    for path_image in filenames:
        image = os.path.abspath(os.path.join(dirpath, path_image))
        with Image.open(image) as img:
            width, height = img.size
            height_list.append(int(height))
            width_list.append(int(width))
a= height_list
b = width_list

print(a)
print("#####################################################################################################")
print("#####################################################################################################")
print("\n\n\n",b)
print("#####################################################################################################")
print("#####################################################################################################")
print("\n\n\n Maximum Height from all images: ",max(height_list))
print("\n\n\n Maximum Width from all images: ",max(width_list))
#             size_images[path_image] = {'width': width, 'heigth': heigth}
# try:
#     geeky_file = open('sizesfile.txt', 'wt')
#     geeky_file.write(str(size_images))
#     geeky_file.close()

# except:
#     print("Unable to write to file")
# print(size_images)
# with open('size_case.json','w') as f:
#     json.dump(size_images,f,ensure_ascii=False)
```
